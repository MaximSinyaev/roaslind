Определение p-distance между строками и построение матрицы

```python
def read_fasta_format(filename):
    d = defaultdict(str)
    with open(filename, 'r') as f:
        name = None
        for line in f:
            if line[0] == '>':
                if name:
                    d[name] = s
                name = line[1:-1]
                s = ''
            else:
                s += line.strip().replace('\n', '')
                print(s)
        d[name] = s
    return d
```

```python
dna_dict = read_fasta_format('./rosalind_pdst.txt')
```

```python
def count_p_distance(s1, s2):
    k = 0
    l = len(s1)
    for i in range(l):
        if s1[i] != s2[i]:
            k += 1
    return k / l
```

```python
for _name, dna in dna_dict.items():
    for _name_2, dna_2 in dna_dict.items():
        if _name == _name_2:
            print(f"{0.0:.5f}", end=' ')
            continue
        print(f"{count_p_distance(dna, dna_2):.5f}", end=' ')
    print()
```
