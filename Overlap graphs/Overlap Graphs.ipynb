```python
from collections import defaultdict
from collections import deque
```

```python
class Graph():
    def __init__(self):
        self.vertex = deque()
        self.edge = 0
        self.adj_list = defaultdict()
        
    def add_vertex(self, name, adj):
        self.vertex.append(name)
        self.adj_list[name] = (adj[:3], adj[-3:])
        
    def find_neighbours(self):
        for name in self.vertex:
            end = self.adj_list[name][1]
            for name_2 in self.vertex:
                if name_2 == name:
                    continue
#                 print(end, self.adj_list[name_2][0])
                if end == self.adj_list[name_2][0]:
                    print(name, name_2)
```
