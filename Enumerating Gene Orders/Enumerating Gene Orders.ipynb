```python
n = int(input())
```

```python
def func(ar, k=0, res=list()):
#     print(ar)
    if len(ar) == 0:
        print()
        return k
    for i, el in enumerate(ar):
        print(el, end=' ')
        res.append(el)
        k = func(ar[0:i] + ar[i + 1:], k + 1, res)
        print(res)
    return k
    
```

```python
func(list(range(1, n + 1)))
```
