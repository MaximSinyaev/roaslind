```python
def func(ar, k=0, res=list()):
    
    if len(ar) == 0:
        print()
        return k
    for i, el in enumerate(ar):
        print(el, end=' ')
        res.append(el)
        k = func(ar[0:i] + ar[i + 1:], k + 1, res)
        print(res)
    return k
    
```

```python
def rec(idx, n, ar, file=sys.stdout):
    if idx == n:
        print(*ar, file=file, flush=False)
        return
    for i in range(1, n + 1):
        if i not in ar:
            ar[idx] = i
            rec(idx + 1, n, ar)
            ar[idx] = 0
```
